```python
# import library
import pandas as pd
from textblob import TextBlob
```

```python
# Load data
data = pd.read_csv('lawyers_california.csv')
```

```python
# Display data
display(data)
```

```python
# Remove unwanted data columns
processed_data = data.drop(['address', 'infoSnippet', 'phone'], axis=1)
```

```python
# Display processed data
display(processed_data)
```

```python
# Check dataset information
processed_data.info()
```

```python
# Check for data duplicates
processed_data.duplicated().sum()
```

```python
# Drop duplicated data
processed_data.drop_duplicates(inplace=True)
```

```python
# Recheck for data duplicates
processed_data.duplicated().sum()
```

```python
# Check for missing values
processed_data.isnull().sum()
```

```python
# Drop rows with missing values except for website
complete_data = processed_data.dropna(subset=['rating', 'reviewSnippet'])
complete_data.reset_index(drop=True, inplace=True)
```

```python
# Display complete data
complete_data
```

```python
# Recheck for missing values
complete_data.isnull().sum()
```

```python
# Use binary indicator for website presence
complete_data['Website_presence'] = complete_data['website'].notnull().astype(int)
```

```python
# Display complete data
complete_data
```

### Sentiment Analysis

```python
# Function that returns sentiment polarity and subjectivity
def calculate_sentiment(text):
    analysis = TextBlob(str(text))
    return analysis.sentiment.polarity, analysis.sentiment.subjectivity

# Apply the function and directly create two new columns in the DataFrame
complete_data[['Review_Sentiment', 'Review_Subjectivity']] = complete_data['reviewSnippet'].apply(
    lambda x: pd.Series(calculate_sentiment(x))
)
```

```python
# Display complete data
display(complete_data)
```

```python
# Save the dataset
complete_data.to_csv('Processed_data.csv', index=False)
```
